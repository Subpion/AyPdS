## analisis y discucion

para poder hallar la funcion tranferencia es nesesaria saber cuanto es la caida de tencion tanto en la entrada como en la salida, para calcularemos la caida de tencion en el la salida a traves del divisor de tencion asumiendo que conocemos la entrada. para ello vamos a calcular la admitancia total en los componentes en serie como en los componentes en derivacion (paralelo).
admitancia en serie
\begin{equation}
y_1=\frac{1}{z_1} =\frac{d}{R} + Cas + \frac{b}{Ls}
\end{equation}
admitancia en derivacion
\begin{equation}
y_2=\frac{1}{z_2} =\frac{1-d}{R} + C(1-a)s + \frac{1-b}{Ls}
\end{equation}
luego expresamos el divisor de tencion en funcion de las admitancias
\begin{equation}
V_2=V_1*\frac{z_2}{z_1+z_2}
\end{equation}

\begin{equation}
T(s)=\frac{V_2}{V_1}=\frac{\frac{1}{y_2}}{\frac{1}{y_1}+\frac{1}{y_2}}
\end{equation}

\begin{equation}
T(s)=\frac{V_2}{V_1}=\frac{1}{\frac{y_1+y_2}{y_1*y_2}}*\frac{1}{y_2}
\end{equation}

\begin{equation}
T(s)=\frac{V_2}{V_1}=\frac{y_1}{y_1+y_2}
\end{equation}

una vez hallada la exprecion de la funcion tranferencia en funcion de s procedemos a simplificarla de manera que quede como cociente de dos polinomios de grado 2

\begin{equation}
T(s)=\frac{V_2}{V_1}=\frac{\frac{d}{R} + Cas + \frac{b}{Ls}}{\frac{d}{R} + Cas + \frac{b}{Ls}+\frac{1-d}{R} + C(1-a)s + \frac{1-b}{Ls}}
\end{equation}


\begin{equation}
T(s)=\frac{V_2}{V_1}=\frac{\frac{d}{R} + Cas + \frac{b}{Ls}}{\frac{1-d+d}{R} + C(1-a+a)s + \frac{1-b+b}{Ls}}
\end{equation}

\begin{equation}
T(s)=\frac{V_2}{V_1}=\frac{\frac{d}{R} + Cas + \frac{b}{Ls}}{\frac{1}{R} + Cs + \frac{1}{Ls}}
\end{equation}

\begin{equation}
T(s)=\frac{V_2}{V_1}=\frac{\frac{s}{C}(\frac{d}{R} + Cas + \frac{b}{Ls})}{\frac{s}{C}(\frac{1}{R} + Cs + \frac{1}{Ls})}
\end{equation}

\begin{equation}
T(s)=\frac{V_2}{V_1}=\frac{\frac{sd}{RC} + as^2 + \frac{b}{CL}}{\frac{s}{CR} + s^2 + \frac{1}{LC}}
\end{equation}

\begin{equation}
T(s)=\frac{V_2}{V_1}=\frac{a(\frac{sd}{aRC} + s^2 + \frac{b}{aCL})}{s^2  + \frac{s}{CR} + \frac{1}{LC}}
\end{equation}

\begin{equation}
T(s)=\frac{V_2}{V_1}=\frac{a(s^2+\frac{s}{RC}\frac{d}{a} + \frac{1}{CL}\frac{b}{a})}{s^2 + \frac{s}{CR}  + \frac{1}{LC}}
\end{equation}
una vez hallada la la funcion tranferencia procedemos a escribir los parametros en funcion de los componenetes del circuito  
\begin{equation}
w_{0p}^2=\frac{1}{LC} \quad y \quad \frac{w_{0p}}{Q_p}=\frac{1}{RC}
\end{equation}

\begin{equation}
w_{0z}^2=\frac{1}{LC}\frac{b}{a} \quad y \quad \frac{w_{0z}}{Q_z}=\frac{1}{RC}\frac{d}{a}
\end{equation}

\begin{equation}
k=a
\end{equation}
y finalmente llegamos a la funcion de tranferencia de \$T(s)$
\begin{equation}
T(s)=\frac{k(s^2+\frac{w_{0z}}{Q_z}s + w_{0z}^2)}{s^2 + \frac{w_{0p}}{Q_p}s + w_{0p}^2}
\end{equation}

\begin{equation}
b=9LC \quad y \quad d=0
\end{equation}
nos queda un sistema de 3 ecuaciones y 4 icognitas por lo que es un sistema compatible indeterminado, por lo que tenemos una relacion de entre los componentes para llegar a las caracteristicas pedidas del filtro 
\begin{equation}
\left \{
\begin{array}{rr}
b=9LC\\
LC=1\\
RC=\sqrt{2}
\end{array}
\right .
\end{equation}
para el analisis de polos y zeros, nesesitamos calcular las raices del polinomio del denominador y la del numerador, de este modo hallamos que hay dos polos en $ 1*e^{\frac{3\pi}{4}}$ y en  $1*e^{\frac{5\pi}{4}}$ por lo que estan en el radio unitario, luego los zeros se ubican en $-3j$ y $3j$ por lo que el diagrama de polos y zeros queda:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

fig, ax = plt.subplots()

circle = plt.Circle((0, 0), 1, color='blue', fill=False, linestyle='--', linewidth=2)


ax.add_artist(circle)
polos=([-np.sqrt(2)/2+np.sqrt(2)*1j/2,-np.sqrt(2)/2-np.sqrt(2)*1j/2])
zeros=([3j,-3j])
plt.figure(1)

plt.plot(np.real(polos[0]), np.imag(polos[0]), 'x', markersize=10, label='Polos',color="r")
plt.plot(np.real(polos[1]), np.imag(polos[1]), 'x', markersize=10,color="r")

plt.plot(np.real(zeros[0]), np.imag(zeros[0]), 'o', markersize=10, label='zeros', fillstyle='none',color="b")
plt.plot(np.real(zeros[1]), np.imag(zeros[1]), 'o', markersize=10,fillstyle='none',color="b")
plt.axhline(0, color='k', lw=0.5)
plt.axvline(0, color='k', lw=0.5)

plt.xlim(-2,1)
plt.ylim(-4,4)
ax.axhline(0, color='black')
ax.axvline(0, color='black')
ax.legend()
plt.grid(True)
plt.show()
```

luego proseguimos a analizar el modulo en respuesta de frecuencia de la funcion tranferencia, para ello evaluamos en $s=jw$ por lo que queda la funcion tranferencia igual a:  
\begin{equation}
T(jw)_1=\frac{(jw)^2+ 9 }{(jw)^2 + jw \sqrt{2}+ 1}
\end{equation}
\begin{equation}
T(jw)_1=\frac{-w^2+ 9 }{-w^2 + jw \sqrt{2}+ 1}
\end{equation}
y tomamos modulos tanto en denominador como numerador
\begin{equation}
|T(jw)_1|=\frac{|-w^2+ 9| }{|-w^2 + jw \sqrt{2}+ 1|}
\end{equation}

\begin{equation}
|T(jw)_1|=\frac{|-w^2+ 9| }{\sqrt{(-w^2+1)^2 + (w \sqrt{2})^2}}
\end{equation}
si graficamos el modulo en funcion de la frecuencia en radianes obtenemos:

```python
w=np.arange(0,100,1)
t_1=((-w**2)+9)/(-w**2+np.sqrt(2)*w*(1j)+1)
plt.plot(w, np.abs(t_1))
plt.title('Respuesta en Magnitud')
plt.xlabel('frecuencia angular [rad/s]')
plt.ylabel('|T(jω)|')
plt.grid(True, which='both', ls=':')
plt.show()
```

luego calulamos la fase 
\begin{equation}
T(jw)_1=\frac{(jw)^2+ 9 }{(jw)^2 + jw \sqrt{2}+ 1}
\end{equation}
\begin{equation}
T(jw)_1=\frac{-w^2+ 9 }{-w^2 + jw \sqrt{2}+ 1}
\end{equation}
\begin{equation}
T(jw)_1=\frac{-w^2+ 9 }{-w^2 + jw \sqrt{2}+ 1}\frac{-w^2 + 1 - jw \sqrt{2}}{-w^2 + 1 - jw \sqrt{2}}
\end{equation}
\begin{equation}
T(jw)_1=\frac{(-w^2+ 9)(-w^2 + 1 - jw \sqrt{2}) }{(-w^2+1)^2 + (w \sqrt{2})^2}
\end{equation}

\begin{equation}
T(jw)_1=\frac{(-w^2+ 9)(-w^2 + 1) - (jw \sqrt{2})(-w^2+ 9) }{(-w^2+1)^2 + (w \sqrt{2})^2}
\end{equation}
\begin{equation}
T(jw)_1=\frac{(-w^2+ 9)(-w^2 + 1)}{(-w^2+1)^2 + (w \sqrt{2})^2}-j\frac{(w \sqrt{2})(-w^2+ 9) }{(-w^2+1)^2 + (w \sqrt{2})^2}
\end{equation}
\begin{equation}
\angle{T(jw)_1}=-tg^{-1}(\frac{w \sqrt{2}}{-w^2 + 1})
\end{equation}
luego lo graficamos y obtenemos que:

```python
plt.plot(w, np.angle(t_1))
plt.title('Fase')
plt.xlabel('frecuencia angular [rad/s]')
plt.ylabel('Fase [°]')
plt.grid(True, which='both', ls=':')
plt.show()
```
