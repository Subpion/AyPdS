## Introducción teórica: Transformada Z

La Transformada Z convierte una señal discreta en una representación en el dominio complejo. Se define como:

$$
X(z) = \sum_{n=-\infty}^{\infty} x(n) z^{-n}
$$

La propiedad fundamental usada en todos los sistemas es el **desplazamiento temporal**:

$$
\mathcal{Z}\{x(n-k)\} = z^{-k}X(z)
$$

Para un sistema LTI, la relación entrada–salida se expresa como:

$$
Y(z) = H(z)X(z)
$$

donde \(H(z)\) es la función de transferencia.


## Respuesta en frecuencia

La respuesta en frecuencia se obtiene evaluando la función de transferencia en el **círculo unitario**:

$$
H(e^{j\omega}) = H(z)\big|_{z = e^{j\omega}}
$$

Esto permite analizar:

- Magnitud:  
  $$|H(e^{j\omega})|$$
- Fase:  
  $$\arg\{H(e^{j\omega})\}$$

Cada retardo en el tiempo se convierte en:

$$
z^{-k}\Big|_{z=e^{j\omega}} = e^{-j\omega k}
$$

lo cual representa un cambio de fase lineal según la frecuencia.

A continuación se muestran las transformadas Z de cada sistema.


## Analisis y discusion

### Transformada Z de cada sistema

Recordemos que:
\begin{aligned}
\mathcal{Z}\{x(n-k)\} = z^{-k}X(z)\\
\mathcal{Z}\{y(n)\} = Y(z)\\
\end{aligned}

#### a) \( y(n) = x(n-3) + x(n-2) + x(n-1) + x(n) \)


\begin{aligned}
Y(z) &= \mathcal{Z}\{x(n-3)\} + \mathcal{Z}\{x(n-2)\} + \mathcal{Z}\{x(n-1)\} + \mathcal{Z}\{x(n)\} \\
Y(z)     &= z^{-3}X(z) + z^{-2}X(z) + z^{-1}X(z) + X(z) \\
Y(z)     &= \big(1 + z^{-1} + z^{-2} + z^{-3}\big)X(z)
\end{aligned}




#### b) \( y(n) = x(n-4) + x(n-3) + x(n-2) + x(n-1) + x(n) \)


\begin{aligned}
Y(z) &= \mathcal{Z}\{x(n-4)\} +\mathcal{Z}\{x(n-3)\} + \mathcal{Z}\{x(n-2)\} + \mathcal{Z}\{x(n-1)\} + \mathcal{Z}\{x(n)\} \\
Y(z) &= z^{-4}X(z) + z^{-3}X(z) + z^{-2}X(z) + z^{-1}X(z) + X(z) \\
Y(z)     &= \big(1 + z^{-1} + z^{-2} + z^{-3} + z^{-4}\big)X(z)
\end{aligned}



#### c) \( y(n) = x(n) - x(n-1) \)


\begin{aligned}
Y(z) &= \mathcal{Z}\{x(n)\} - \mathcal{Z}\{x(n-1)\} \\
Y(z)     &= X(z) - z^{-1}X(z) \\
Y(z)     &= \big(1 - z^{-1}\big)X(z)
\end{aligned}




#### d) \( y(n) = x(n) - x(n-2) \)


\begin{aligned}
Y(z) &= \mathcal{Z}\{x(n)\} - \mathcal{Z}\{x(n-2)\} \\
Y(z) &= X(z) - z^{-2}X(z) \\
Y(z) &= \big(1 - z^{-2}\big)X(z)
\end{aligned}



### Función de transferencia \(H(z)\)

La **función de transferencia** se define como: 

\begin{aligned}
 H(z)=\dfrac{Y(z)}{X(z)}  (con X(z)\neq 0)\\
\end{aligned}



#### a) y(n)=x(n-3)+x(n-2)+x(n-1)+x(n) 


\begin{aligned}
Y(z) &= \big(1+z^{-1}+z^{-2}+z^{-3}\big)\,X(z) \\
H(z) &= \frac{Y(z)}{X(z)} = 1+z^{-1}+z^{-2}+z^{-3}\\
H(z) &= \dfrac{z^3(1 + z^{-1} + z^{-2} + z^{-3})}{z^3}\\
H(z) &= \dfrac{z^3 + z^2 + z + 1}{z^3}\\
\end{aligned}



#### b)  y(n)=x(n-4)+x(n-3)+x(n-2)+x(n-1)+x(n) 

\begin{aligned}
Y(z) &= \big(1+z^{-1}+z^{-2}+z^{-3}+z^{-4}\big)\,X(z) \\
 H(z) &= \frac{Y(z)}{X(z)} = 1+z^{-1}+z^{-2}+z^{-3}+z^{-4}\\
H(z)&= \frac{z^4(1 + z^{-1} + z^{-2} + z^{-3} + z^{-4})}{z^4}\\
H(z)     &= \frac{z^4 + z^3 + z^2 + z + 1}{z^4}\\
\end{aligned}



#### c)  y(n)=x(n)-x(n-1) 

\begin{aligned}
Y(z) &= \big(1 - z^{-1}\big)\,X(z) \\
H(z) &= \frac{Y(z)}{X(z)} = 1 - z^{-1}\\
H(z) &= \frac{z(1 - z^{-1})}{z}\\
H(z) &= \frac{z - 1}{z}\\
\end{aligned}


#### d)  y(n)=x(n)-x(n-2) 

\begin{aligned}
Y(z) &= \big(1 - z^{-2}\big)\,X(z) \\
H(z) &= \frac{Y(z)}{X(z)} = 1 - z^{-2}\\
H(z) &= \frac{z^2(1 - z^{-2})}{z^2}\\
H(z) &= \frac{z^2 - 1}{z^2}\\
\end{aligned}

### Respuesta en frecuencia: módulo y fase

Recordemos que la respuesta en frecuencia se obtiene evaluando en el círculo unitario:

$$
H(e^{j\omega}) = H(z)\big|_{z = e^{j\omega}}
$$



#### a) \( y(n) = x(n-3) + x(n-2) + x(n-1) + x(n) \)

Función de transferencia:

$$
H_a(z) = 1 + z^{-1} + z^{-2} + z^{-3}
$$

Respuesta en frecuencia:

$$
H_a(e^{j\omega}) = 1 + e^{-j\omega} + e^{-j2\omega} + e^{-j3\omega}
$$

Usando la identidad del sumador de exponentiales:

$$
H_a(e^{j\omega}) = e^{-j\frac{3\omega}{2}} \, \frac{\sin(2\omega)}{\sin(\omega/2)}
$$

**Módulo:**

$$
|H_a(e^{j\omega})| = \left|\frac{\sin(2\omega)}{\sin(\omega/2)}\right|
$$

**Fase:**

$$
\angle H_a(e^{j\omega}) = -\frac{3\omega}{2}
\quad (\text{con saltos donde cambia el signo})
$$



#### b) \( y(n) = x(n-4) + x(n-3) + x(n-2) + x(n-1) + x(n) \)

$$
H_b(z) = 1 + z^{-1} + z^{-2} + z^{-3} + z^{-4}
$$

$$
H_b(e^{j\omega}) = 1 + e^{-j\omega} + e^{-j2\omega} + e^{-j3\omega} + e^{-j4\omega}
$$

$$
H_b(e^{j\omega}) = e^{-j2\omega} \, \frac{\sin\left(\frac{5\omega}{2}\right)}{\sin(\omega/2)}
$$

**Módulo:**

$$
|H_b(e^{j\omega})| = \left|\frac{\sin\left(\frac{5\omega}{2}\right)}{\sin(\omega/2)}\right|
$$

**Fase:**

$$
\angle H_b(e^{j\omega}) = -2\omega
\quad (\text{más saltos de fase por el signo})
$$



#### c) \( y(n) = x(n) - x(n-1) \)

$$
H_c(z) = 1 - z^{-1}
$$

$$
H_c(e^{j\omega}) = 1 - e^{-j\omega}
$$

$$
H_c(e^{j\omega}) = e^{-j\frac{\omega}{2}} \cdot 2j \sin\left(\frac{\omega}{2}\right)
$$

**Módulo:**

$$
|H_c(e^{j\omega})| = 2\left|\sin\left(\frac{\omega}{2}\right)\right|
$$

**Fase:**

$$
\angle H_c(e^{j\omega}) = -\frac{\omega}{2} + \frac{\pi}{2}
$$



#### d) \( y(n) = x(n) - x(n-2) \)

$$
H_d(z) = 1 - z^{-2}
$$

$$
H_d(e^{j\omega}) = 1 - e^{-j2\omega}
$$

$$
H_d(e^{j\omega}) = e^{-j\omega} \cdot 2j \sin(\omega)
$$

**Módulo:**

$$
|H_d(e^{j\omega})| = 2|\sin(\omega)|
$$

**Fase:**

$$
\angle H_d(e^{j\omega}) = -\omega + \frac{\pi}{2}
$$


### Grafico de la ecuacion a)

```python
import numpy as np
import matplotlib.pyplot as plt


w = np.linspace(0, np.pi, 1024)

# Funcion de transferencia H(e^{jw})
H = 1 + np.exp(-1j*w) + np.exp(-2j*w) + np.exp(-3j*w)


H_mag = np.abs(H)
H_phase = np.angle(H)


plt.figure()


plt.subplot(2, 1, 1)
plt.plot(w, H_mag)
plt.title('Magnitud de H(e^{jw})')
plt.ylabel('|H|')
plt.xlabel('Frecuencia (rad/muestra)')
plt.grid(True)


plt.subplot(2, 1, 2)
plt.plot(w, H_phase)
plt.title('Fase de H')
plt.ylabel('fase[rad]')
plt.xlabel('Frecuencia (rad/muestra)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Análisis y Discusión

El análisis de las respuestas en frecuencia obtenidas para los cuatro sistemas permite identificar con claridad la naturaleza de cada uno y relacionar su comportamiento espectral con la estructura de sus ecuaciones en diferencias.

En primer lugar, los sistemas (a) y (b), basados en sumas de muestras retrasadas, presentan un módulo con valores máximos en bajas frecuencias y una marcada atenuación a medida que la frecuencia se acerca a $( \pi )$. Este comportamiento es característico de un filtro pasa-bajos. La razón es que señales de variación lenta tienen valores consecutivos muy similares, y la suma refuerza estas componentes. Por el contrario, en frecuencias altas, los términos $e^{-j\omega k}$ alternan de signo y se producen cancelaciones. Además, la fase decrece linealmente debido a la presencia de retardos $z^{-k}$, lo cual introduce un término proporcional a $-k\omega$. El sistema (b), al incluir un retardo adicional, exhibe una pendiente de fase mayor y mayor selectividad espectral.

Por otro lado, los sistemas (c) y (d) implementan restas de muestras desplazadas, lo que conduce a una magnitud nula en bajas frecuencias. Esto concuerda con la interpretación de un filtro diferenciador, donde se suprimen las componentes lentas porque $x(n)\approx x(n-k)$ cuando la señal es suave. Las componentes de alta frecuencia, en cambio, se amplifican debido a la diferencia pronunciada entre muestras consecutivas. Del mismo modo, la fase muestra un comportamiento lineal, desplazándose según la cantidad de muestras atrasadas involucradas.

En conjunto, las representaciones gráficas verifican las propiedades teóricas de la Transformada Z y muestran que la estructura algebraica de cada $H(z)$ se traduce directamente en un comportamiento espectral coherente y fácilmente interpretable.


## Conclusion


Los resultados obtenidos permiten concluir que las funciones de transferencia analizadas se comportan exactamente como predice la teoría de sistemas LTI en tiempo discreto. Los filtros compuestos por sumas de retardos actúan como pasa-bajos, reforzando componentes de baja frecuencia y atenuando variaciones rápidas. En contraste, los filtros definidos por diferencias implementan un comportamiento pasa-altos, suprimiendo señales lentas y destacando cambios abruptos.

Las gráficas de magnitud y fase confirman estas características: los módulos presentan máximos y mínimos en las frecuencias esperadas y las fases siguen patrones lineales asociados a los retardos presentes en los términos $z^{-k}$. Esto valida tanto la formulación de las funciones de transferencia como la implementación computacional utilizada para evaluar $H(e^{j\omega})$.
